### 1.Install libraries, data downloading 

```python
# install libraries
import pandas as pd
import numpy as np
import glob
```

```python
# was taken a dataset from kaggle
data = pd.read_csv('healthcare-dataset-stroke-data.csv')
data.head()
```

```python
# checking information
data.info()
```

### 2.Making a list of files (110)

```python
#distribute the records into 110 files (according to the technical specifications)
files = [] # make empty list

# create a loop for making 110 files in csv format
for i in range(1, 111): #determine the number of records from 1 to 111 (110 in total)
    file_num = str(i) # declare a variable as a "string"
    if len(file_num) == 1: # if the file name is 1 character long
        file_num = '00' + file_num # add 00 to file name
    elif len(file_num) == 2: # if the file name is 2 character long (example 12)
        file_num = '0' + file_num # add 0 to file name
    files.append(f'data_000000000{file_num}.csv') # using append method for adding main part of the file name - data_000000000
```

```python
# split dataset for a small parts
a = np.array_split(data, len(files))
```

```python
# create a loop for saving files in a directory
for file, d in zip(files, a):
    d.to_csv('data\\' + file)
```

```python
# determinea variable 'files_list'
files_list = glob.glob('data\*.csv')

# checking
files_list
```

### 3.Combined 110 files back to 1 dataset

```python
# determinea variable 'combined'
combined = pd.DataFrame() # first make empty dataframe

# make a loop which will check each file, read it and add a colunm 'filename'
for file in files_list:
    data = pd.read_csv(file, index_col=[0])
    data['filename'] = file
    combined = pd.concat([combined, data])

combined.reset_index(drop=True, inplace=True)
```

```python
# checking the result 
combined[:5]
```

```python
# write a single file .csv
combined.to_csv('combined.csv',
                sep = ';') # will allow to split data into columns in excel (if will be nessesary)
```
